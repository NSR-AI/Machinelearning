## plotting the data

we can plot the data between these points using the scatter() fucntion in the matlplotlib library 

Arguments marker is used to show the points and c to show the color of the marker


```python
plt.scatter(x_train,y_train,marker='o',c='b')
plt.title("Housing prices")
plt.ylabel("prices (in 1000s of dollars)")
plt.show()
```

```python
w=100
b=100
print(f"w:{w}")
print(f"b:{b}")
```

```python
def model(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
    return f_wb
```

```python
temp_f_wb=model(x_train,w,b)
plt.plot(x_train,temp_f_wb,c='b',label='Our prediction')
plt.scatter(x_train,y_train,marker='x',c='r')
```

# Using Sklearn (Other type of implementation)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("C:\\Users\\nshra\\Machine learning\\supervisedmachine learning\\homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price(USD$)")
plt.scatter(df.area,df.price,color='red',marker='*')
plt.show()
```

```python
x=df['area'].values.reshape(-1,1)
y=df['price']
reg=linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
135.78767*3300+180616
```

```python
d=pd.read_csv("C:\\Users\\nshra\\Machine learning\\supervisedmachine learning\\areas.csv")
d
```

```python
p=reg.predict(d)
```

```python
d['prices']=p
d
```

```python
d.to_csv("prediction.csv",index=False)
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("price(USD$)")
plt.scatter(df.area,df.price,color='red',marker='*')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
plt.show()
```
